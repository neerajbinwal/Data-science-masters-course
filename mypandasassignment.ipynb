```python
df3= pd.DataFrame({"Values":[1,2,3,4,5,6,9,10]})
```

```python
df3["Values"].mean()
```

```python
def my_maths(x):
    mean= x["Values"].mean()
    median= x["Values"].median()
    standard_deviation=x["Values"].var()
    print(f"""    The mean is {mean},
    The median is {median},
    The standard deviation is {standard_deviation}""")
```

```python
my_maths(df3)
```
