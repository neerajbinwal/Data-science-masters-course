Ans 3.ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

OverFlowError
ZeroDivisionError
FloatingPointError
These errors are all capable of crashing a code in Python. It is essential to catch an error because you do not want your code to crash as a result of incorrect input from you or a user.



```python
###we will give example of zerodivison error 
a=11
try:
    c=a/0
except ZeroDivisionError as e:
    print(e)
```

```python
#example of over flow errot
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
```
