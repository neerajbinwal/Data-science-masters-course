Ans 2

```python
import numpy as np
import matplotlib.pyplot as plot
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plot.Figure(figsize=(6,4))
plot.scatter(x,y)
plot.xlabel("this is my x axis")
plot.ylabel("this is my y axis")
plot.title("MY scatter plot")
```

The matplotlib.pyplot.subplots method provides a way to plot multiple plots on a single figure

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plot.subplot()
plot.plot(x1,y1)
plot.subplot()
plot.plot(x2,y2)
plot.subplot()
plot.plot(x3,y3)
plot.subplot()
plot.plot(x4,y4)
```

Ans4 A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. 

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plot.barh(company,profit,color='b')
plot.title("MY BAR PLOT")
plot.xlabel("Companies")
plot.ylabel("Profit")
```
