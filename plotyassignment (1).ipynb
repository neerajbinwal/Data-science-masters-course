Displots in Seaborn are univariant sets of collected data used to visualize the univariate or bivariate distribution of data in histogram form.


```python


from plotly.figure_factory import create_distplot
import numpy as np

np.random.seed(1)
x=np.random.randn(100)
hist_data =[x]
group_labels = ['distplot basic example']
fig = create_distplot(hist_data, group_labels)
fig.show()
```
